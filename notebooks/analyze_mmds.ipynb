```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
model_paths = {
    'RNN-SW': '../runs/RNN/SlidingWindow_noshift_lr1e-5_2024-04-26_22-16-45',
    'RNN-SW + 0.2Shift': '../runs/RNN/SlidingWindow_Shift0.2_lr1e-5_2024-04-26_18-19-36',
    'RNN-SW + 0.8Shift': '../runs/RNN/SlidingWindow_Shift0.8_lr1e-5_2024-04-26_16-28-45',
    'RNN-SW + 0.9Shift': '../runs/RNN/data_aug_0.9_all_tones',
    'RNN-noaug': '../runs/RNN/Without_Aug_2024-04-25_14-39-15',
}
```

```python
for model_name, model_path in model_paths.items():
    df = pd.read_csv(f'{model_path}/logs.csv')
    plt.plot(df['epoch'], df['mmd_notes'], label=model_name)
plt.grid()
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('MMD Notes')
plt.title('MMD of notes comparing various\ndata augmentation techniques on RNN model')
plt.show()
```
